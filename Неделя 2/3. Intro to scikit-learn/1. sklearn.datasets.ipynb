make_circles позволяет получить датасет, который выглядит как две вложенные окружности

```python
circles = datasets.make_circles()
```

```python
print('features: {}'.format(circles[0][:10])) # координаты
print('target: {}'.format(circles[1][:10]))   # метки классов
```

Для визуализации импортируем объект ListedColormap

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['red', 'yellow'])
pyplot.figure(figsize(8, 8))
pyplot.scatter(list(map(lambda x: x[0], circles[0])), 
               list(map(lambda x: x[1], circles[0])),
               c=circles[1], 
               cmap=colors)
```

```python
def plot_2d_dataset(data, colors):
    pyplot.figure(figsize(8, 8))
    pyplot.scatter(list(map(lambda x: x[0], data[0])),
                  list(map(lambda x: x[1], data[0])),
                  c=data[1],
                  cmap=colors)
```

Добавим небольшой шум в данные

```python
noisy_circles = datasets.make_circles(noise=0.05)
```

```python
plot_2d_dataset(noisy_circles, colors)
```

#### datasets.make_classification 

```python
simple_classification_problem = datasets.make_classification(n_features=3,
                                                            n_informative=2,
                                                            n_redundant=1,
                                                            n_clusters_per_class=1,
                                                            random_state=1)
```

```python
plot_2d_dataset(simple_classification_problem, colors)
```

```python
classification_problem = datasets.make_classification(n_features=2,
                                                     n_informative=2,
                                                     n_classes=4,
                                                     n_redundant=0,
                                                     n_clusters_per_class=1,
                                                     random_state=1)
```

```python
plot_2d_dataset(classification_problem, colors)
```

### Визуализация выборки 

```python
from pandas import DataFrame
```

```python
iris_frame = DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
```

```python
iris_frame.head()
```

```python
iris_frame['target'] = iris_frame['target'].apply(lambda x: iris.target_names[x])
```

```python
iris_frame.head()
```

```python
iris_frame[iris_frame.target == 'setosa'].hist('sepal length (cm)')
```

```python
pyplot.figure(figsize(20, 24))
plot_number = 0
for feature_name in iris.feature_names:
    for target_name in iris.target_names:
        plot_number += 1
        pyplot.subplot(4, 3, plot_number)
        pyplot.hist(iris_frame[iris_frame.target == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel('cm')
        pyplot.ylabel(feature_name[:-4])
```
